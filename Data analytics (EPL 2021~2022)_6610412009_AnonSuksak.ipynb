## 2.2 Did obtaining yellow and red cards affect the home team's win rate?

## Step 1 กำหนดให้ H = 1, และ D, A = 0 เพื่อจะวิเคราะห์โอกาสการชนะ และทำ groupby เพื่อดูข้อมูล

```python
df2['HomeWin'] = df2['FTR'].apply(lambda result: 1 if result == 'H' else 0)
```

```python
df2
```

```python
win_ = df2.groupby("HomeTeam")[["HY","HR","HomeWin"]].sum()

win_.sort_values(by=["HY", "HR"], ascending=False) 

#HY = Hometeam yellow card
#HR = Hometeam red card
```

## Step 2 วิเคราะห์ Regression โดยต้องการดูว่า การได้รับใบเหลือง หรือใบแดง หรือทั้งสองอย่าง จะกระทบต่ออัตราการชนะของทีมเจ้าบ้านหรือไม่
## Model: lnHometeamWin = beta0 + beta1HY + beta2HR

```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
```

```python
# Create the dependent variable (y) and independent variables (X)
y = np.log1p(df2['HomeWin'])
X = df2[['HY', 'HR']]

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression results
print(model.summary())
```

## อีกโค้ด ที่ใช้ Sklearn

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd

# Assuming df2 is your DataFrame
X = df2[['HY', 'HR']]
Y = np.log(df2['HomeWin'] + 1)  # Adding 1 to handle cases where 'HomeWin' is zero

# Create a linear regression model
model = LinearRegression()

# Fit the model
model.fit(X, Y)

# Print the coefficients
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```

## Step 3 ผลการวิเคราะห์ 
## logHometeamwin = 0.3882 - 0.0479HY - 0.2338HR
- ผลการศึกษา พบว่า ค่าสัมประสิทธิ์ของทั้ง HR (Hometeam red card), HY (Hometeam yellow card) มีค่าติดลบ ดังนั้นหาก obtaining yellow and red cards จะทำให้โอกาสในการชนะลดลง
- เช่น หาก Hometeam ได้รับ Red card 0 ใบ Yellow card 0 ใบ โอกาสชนะจะเท่ากับ HomeWin = 0.3882-0.0479(0) - 0.2338(0) = 0.3882%
- แต่หากได้รับ Red card 1 ใบ Yellow card 1 ใบ โอกาสชนะจะลดลงเป็น HomeWin = 0.3882-0.0479(1) - 0.2338(1) = 0.1065%
