```python
import pandas as pd
```

## Step 1 get data

```python
df = pd.read_csv('https://raw.githubusercontent.com/Anonsuksak/DADS5001_Plotly/main/Apple%E2%80%99s%20Product%20Revenue%20(2007-2023).csv')
df
```

- data source: https://www.visualcapitalist.com/charted-apples-product-revenue-2007-2023/

## Step 2 data cleansing

```python
columns_to_clean = ['iPhone', 'Accessories', 'Mac', 'iPad', 'iPod', 'Services']

for column in columns_to_clean:
    df[column] = df[column].str.replace("$", "").str.replace("B", "").str.replace("-", "0.00")
```

- ทำ data cleansing เนื่องจาก raw data ยังติดค่า $, B, - ต้องทำการลบเครื่องหมายเหล่านี้ออก

```python
df2 = df.set_index("Apple Revenue")
df2
```

```python
df2.info()
```

- หลังจากตรวจสอบข้อมูล พบว่า ประเภทข้อมูลเป็น object ทำให้ไม่สามารถคำนวณทางคณิตศาสตร์ได้ จึงแปลง object ให้เป็น float เพื่อที่จะสามารถหาสัดส่วนรายได้แต่ละประเภทสินค้าต่อรายได้ทั้งหมดได้

```python
columns_to_clean = ['iPhone', 'Accessories', 'Mac', 'iPad', 'iPod', 'Services']

for column in columns_to_clean:
    df2[column] = df2[column].astype(float)
```

- ต่อมาจะเป็นการหา total revenue เพื่อที่จะหาสัดส่วนรายได้แต่ละประเภทสินค้าได้

```python
df2['Total Revenue'] = df2[['iPhone', 'Accessories', 'Mac', 'iPad', 'iPod', 'Services']].sum(axis=1)
df2
```

## Step 3 data calculation

```python
columns_to_clean = ['iPhone', 'Accessories', 'Mac', 'iPad', 'iPod', 'Services','Total Revenue']

for column in columns_to_clean:
    df2[column] = ((df2[column]/df2['Total Revenue'])*100).round(1)
```

```python
df2['Services & other'] = (df2['Accessories'] + df2['Services']).round(1)
df2.head()
```

```python
df3 = df2.loc['2008':'2022',['iPhone','Mac','iPad','iPod','Services & other']]
df3.head()
```

```python
df4 = df.iloc[1:16,:]
df4
```

```python
import plotly.graph_objects as go

title = 'Apple’s Revenue by Product (2008-2022)'
labels = ['iPhone', 'Mac', 'iPad', 'iPod', 'Services & other']
colors = ['rgb(67,67,67)', 'rgb(115,115,115)', 'rgb(189,189,189)', 'rgb(255,0,0)', 'rgb(49,130,189)']

mode_size = [12, 8, 8, 8, 8]
line_size = [4, 2, 2, 2, 2]

x_data = df3.index # ใช้ index เป็นแกน x
y_data = df3[['iPhone', 'Mac', 'iPad', 'iPod', 'Services & other']]

fig = go.Figure()

for i in range(0, 5):
    fig.add_trace(go.Scatter(x=x_data, y=y_data[labels[i]], mode='lines',
        name=labels[i],
        line=dict(color=colors[i], width=line_size[i]),
        connectgaps=True,
    ))

    # endpoints
    fig.add_trace(go.Scatter(
        x=[x_data[0], x_data[-1]],
        y=[y_data.iloc[0, i], y_data.iloc[-1, i]],  # เปลี่ยนเป็น y_data.iloc[0, i] เป็น y_data.iloc[0][i]
        mode='markers',
        marker=dict(color=colors[i], size=mode_size[i])
    ))

fig.update_layout(
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)',
        ),
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=False,
        showticklabels=False,
    ),
    autosize=False,
    margin=dict(
        autoexpand=False,
        l=100,
        r=20,
        t=110,
    ),
    showlegend=False,
    plot_bgcolor='white'
)

annotations = []

# Adding labels
for label, color in zip(labels, colors):
    # labeling the left_side of the plot
    annotations.append(dict(xref='paper', x=0.05, y=y_data.iloc[0][labels.index(label)],  # เปลี่ยนเป็น y_data.iloc[0][labels.index(label)]
                                  xanchor='right', yanchor='middle',
                                  text=label + ' {}%'.format(y_data.iloc[0][labels.index(label)]),  # เปลี่ยนเป็น y_data.iloc[0][labels.index(label)]
                                  font=dict(family='Arial',
                                            size=12),
                                  showarrow=False))
    # labeling the right_side of the plot
    annotations.append(dict(xref='paper', x=0.95, y=y_data.iloc[-1][labels.index(label)],  # เปลี่ยนเป็น y_data.iloc[-1][labels.index(label)]
                                  xanchor='left', yanchor='middle',
                                  text='{}%'.format(y_data.iloc[-1][labels.index(label)]),  # เปลี่ยนเป็น y_data.iloc[-1][labels.index(label)]
                                  font=dict(family='Arial',
                                            size=12),
                                  showarrow=False))

# Title
annotations.append(dict(xref='paper', yref='paper', x=0.5, y=1.05,
                              xanchor='center', yanchor='bottom',
                              text=title,
                              font=dict(family='Arial',
                                        size=30,
                                        color='rgb(37,37,37)'),
                              showarrow=False))

# Source
annotations.append(dict(xref='paper', yref='paper', x=0.5, y=-0.1,
                              xanchor='center', yanchor='top',
                              text='Source: Apple inc., & visualcapitalist',
                              font=dict(family='Arial',
                                        size=12,
                                        color='rgb(150,150,150)'),
                              showarrow=False))

fig.update_layout(annotations=annotations)

fig.show()
```
